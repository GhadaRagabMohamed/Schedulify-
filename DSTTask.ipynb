```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
import itertools
from collections import Counter
```

```python
# load the file
file_path="studentcourses.xlsx"
xls=pd.ExcelFile(file_path)
```

```python
# Dictionary to store all schedules
sheets_data={}
```

```python
def get_restricted_pairs(df):
    student_courses=df.groupby("اسم الطالب")["اسم المقرر"].apply(list)
    pairs=[]
    for courses in student_courses: 
        pairs.extend(itertools.combinations(courses,2))
    pair_counts=Counter(pairs)
    restricted_pairs=[pair for pair,count in pair_counts.most_common(10)]
    return restricted_pairs
```

```python
exam_schedules=[]
# process each sheet seperately
for sheet_name in xls.sheet_names:
    print(f"processing department:{sheet_name}")
    # read data
    df=pd.read_excel(xls,sheet_name=sheet_name)
    # handle 'علوم طبية' sheet &&"حاسب الي"sheet
    if sheet_name.strip()=="علوم طبية".strip():
        df.columns=["م","رقم الطالب","اسم الطالب","التخصص","رقم المقرر","اسم المقرر","رمز المقرر","رقم الشعبة"]
    elif sheet_name.strip()=="حاسب الي".strip():
        df.columns=["م","رقم الطالب","اسم الطالب","التخصص","رقم المقرر","اسم المقرر","رمز المقرر","رقم الشعبة"]
        df=df[1:].reset_index(drop=True)
    # keep only relevant columns
    df=df[["اسم الطالب","اسم المقرر","رمز المقرر"]]
    print(df.columns)
    print(df.head(3))
    # Standarize data format
    df["اسم الطالب"]=df["اسم الطالب"].astype(str).str.strip().str.lower()
    df["اسم المقرر"]=df["اسم المقرر"].astype(str).str.strip().str.lower()
    df["رمز المقرر"]=df["رمز المقرر"].astype(str).str.strip().str.lower()
    # check for missing value
    missing_values=df.isnull().sum()
    print(f"missing_values in{sheet_name}:\n{missing_values}\n\n")
    # check for duplicate
    duplicate_count=df.duplicated().sum()
    print(f"duplicate_count in{sheet_name}:\n{duplicate_count}\n\n")
    df.drop_duplicates(inplace=True)
    duplicate_count2=df.duplicated().sum()
    print(f"duplicate_count2 in{sheet_name}:\n{duplicate_count2}\n\n")
    print(df.info())
    # Count the number of students in each course
    course_counts=df.groupby("اسم المقرر")["اسم الطالب"].nunique().to_dict() 
    # map course codes to course name
    course_code_map=df.set_index("اسم المقرر")["رمز المقرر"].to_dict()
    restricted_pairs=get_restricted_pairs(df)
    print(f"Courses that should be scheduled seperately:{restricted_pairs}")
    # creat the graph
    G=nx.Graph()
    # add courses as nodes
    courses=df["اسم المقرر"].unique()
    G.add_nodes_from(courses)
    # add edges between courses taken by same student
    for student,group in df.groupby("اسم الطالب"):
        courses_taken=list(group["اسم المقرر"])
        for i in range(len(courses_taken)):
            for j in range(i+1,len(courses_taken)):
                G.add_edge(courses_taken[i],courses_taken[j])
    # calculate degree for each course
    course_degrees=dict(G.degree())
    print(f"\ncourse_degrees : {course_degrees}\n")
    # General statistics
    num_courses=len(G.nodes)
    num_edges=len(G.edges)
    avg_degree=sum(course_degrees.values()) / num_courses
    # display basic statistics
    print(f"Number of courses: {num_courses}")
    print(f"Number of course relationships: {num_edges}")
    print(f"Averge course degree: {avg_degree:.2f}")
    # apply the greedy  coloring algorithms
    coloring=nx.coloring.greedy_color(G,strategy="largest_first")
    # schedule exams based on assigned colors
    schedule={}
    for course,color in coloring.items():
        if color not in schedule:
            schedule[color]=[]
        schedule[color].append(course)
    # distrubute exams across days while considering constraints
    max_exams_per_day=3
    max_exams_per_period=2
    exam_days=[]
    current_day=[]
    current_period=[]
    for color,courses in sorted(schedule.items()):
        for course in courses:
            if len(current_period)<max_exams_per_period and all((c1,course) not in restricted_pairs for c1 in current_period):
                current_period.append(course)
            else:
                if len (current_day)< max_exams_per_day:
                    current_day.append(current_period)
                    current_period=[course]
                else:
                    exam_days.append(current_day)
                    current_day=[current_period]
                    current_period=[course]
    if current_period:
        current_day.append(current_period)
    if current_day:
        exam_days.append(current_day)
    formatted_schedule=[]
    for day_num,periods in enumerate(exam_days,start=1):
        for period_num,courses in enumerate(periods,start=1):
            for course in courses:
                formatted_schedule.append({"Day":day_num,"Period":period_num,"Courses":course,"Student Count":course_counts[course],
                                       "Student Code":course_code_map[course]})
    schedule_df=pd.DataFrame(formatted_schedule)
    exam_schedules.append((sheet_name,schedule_df))
    # visualize the course graph
    plt.figure(figsize=(12,8))
    nx.draw(G,with_labels=True,node_color=list(coloring.values()),cmap=plt.cm.rainbow)
    plt.title("Course Relationship Graph")
    plt.show()
    plt.figure(figsize=(10,6))
    sns.histplot(list(course_degrees.values()),bins=10,kde=True)
    plt.xlabel("Course Degree")
    plt.ylabel("Frequency")
    plt.title("Course Degree Distribution in the Network")
    plt.show()
with pd.ExcelWriter("Exam_Schedule_with_periods.xlsx") as writer:
    for department,df in exam_schedules:
        df.to_excel(writer,sheet_name=department,index=False)
print("Final exam schedule successufuly created: Exam_Schedule_with_periods.xlsx")
  
```
